# Bernstein-Vazirani Algorithm  
The berstein-Vazirani(BV) algorithm demonstrates an exponential speedup over classical methods for the paricular task:  

Supplose we have an oracle that implements the function $f:\{0,1\}^n \rightarrow \{0, 1\}$  

$f(x)$ is promised to be the dot product b/w x and a secret n-bit binary string $s$:  

$f(x) = x\bullet s(mod 2) = (x_1 s_1 + ... + x_n x_n)(mod2) $  

Our goal is to find the secrt string s

# Classical case  
For the case of n = 3, consider s = 101,  

We have access to the function: $f(x) = x\bullet s (mod 2) = (x_1s_1 + x_2s_2 + x_3s_3) (mod 2)$  
In the classical case, we see that the secret string $s$ can be calculated in 3 queries. More generally speaking, it would take n queries to generate enough information to determine $s$

# Quantum case  
The BV altorithm can take advantage of superposition and entanglement to encdoe the problem and produce an ansswer with a single oracle (black box function) query. The setup is a register of n qubits in the $|0\rangle$ state and a single auxiliary qubit in the #|1\rangle$ state. The auxiliary qubit enables the phase kickback. Consider the register of $n$ qubits initialized here:  
$$|0\rangle$$
Application of a Hadamard gate to each qubit results in:
$$H|0\rangle = \frac{1}{\sqrt{2^n}} \sum_x^{n-1}|x\rangle$$
Next, the oracle, $U_f$, performs maps $|x \rangle$ to $(-1)^f(x) |x\rangle$ resulting in:
$$ U_f \frac{1}{\sqrt{2^n}} \sum_x^{n-1}|x\rangle = \frac{1}{\sqrt{2^n}} \sum_x^{n-1} (-1)^{f(x)} |x\rangle = \frac{1}{\sqrt{2^n}} \sum_x^{n-1} (-1)^{a*x}|x\rangle $$
Applying Hadamard gates again returns the following state:
$$ H \frac{1}{\sqrt{2^n}} \sum_x^{n-1} (-1)^{a*x}|x\rangle = \frac{1}{\sqrt{2^n}} \sum^{n-1}_{x,y} (-1)^{a*x} (-1)^{y*x} |y\rangle = \frac{1}{\sqrt{2^n}} \sum^{n-1}_y \sum^{n-1}_x (-1)^{(a\oplus y)*x} |y\rangle$$
In the case of a != y, the entire term becomes zero, In the case offa= y, the state $|a\rangle$ is returned with an amplitude of 1. This means there is a 100% changce of the measurement resulting being the hidden bitstring $a$ with only one call to the oracle!
